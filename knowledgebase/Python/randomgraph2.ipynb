```python
%matplotlib inline
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

np.random.seed(42)
n = 200
departments = ['Engineering', 'Marketing', 'Sales', 'HR', 'Finance']
df = pd.DataFrame({
    'Name':         [f'Employee_{i}' for i in range(1, n+1)],
    'Department':   np.random.choice(departments, n),
    'Age':          np.random.randint(22, 61, n),
    'Salary':       np.round(np.random.uniform(35000, 150000, n), 2),
    'Years_Exp':    np.round(np.random.uniform(0.5, 20, n), 1),
    'Performance':  np.random.randint(1, 6, n),
    'Satisfaction': np.round(np.random.uniform(1, 10, n), 1),
})

out = widgets.Output()

def show_summary(department):
    with out:
        clear_output(wait=True)
        subset = df if department == 'All' else df[df['Department'] == department]
        num_cols = ['Age', 'Salary', 'Years_Exp', 'Performance', 'Satisfaction']
        summary = subset[num_cols].agg(['count', 'mean', 'std', 'min', 'max']).T
        summary.columns = ['Count', 'Mean', 'Std Dev', 'Min', 'Max']
        summary['Mean']    = summary['Mean'].round(2)
        summary['Std Dev'] = summary['Std Dev'].round(2)
        summary['Min']     = summary['Min'].round(2)
        summary['Max']     = summary['Max'].round(2)
        summary['Count']   = summary['Count'].astype(int)

        fmt = {}
        for col in ['Mean', 'Std Dev', 'Min', 'Max']:
            fmt[col] = lambda v: f'{v:,.0f}'

        print('Department: ' + department + '  |  Employees: ' + str(len(subset)))
        print()
        display(summary.style
            .set_table_styles([
                {'selector': 'th', 'props': [('background-color','#4c72b0'),('color','white'),('font-weight','bold'),('padding','8px 12px')]},
                {'selector': 'td', 'props': [('padding','6px 12px'),('border-bottom','1px solid #ddd')]},
                {'selector': 'tr:hover td', 'props': [('background-color','#f0f4ff')]},
            ])
            .format(subset=pd.IndexSlice[['Age','Years_Exp','Performance','Satisfaction'], ['Mean','Std Dev','Min','Max']], formatter='{:.2f}')
            .format(subset=pd.IndexSlice[['Salary'], ['Mean','Std Dev','Min','Max']], formatter='{:,.0f}')
            .background_gradient(subset=pd.IndexSlice[:, 'Mean'], cmap='Blues')
        )

dropdown = widgets.Dropdown(
    options=['All'] + departments,
    value='All',
    description='Department:',
    layout=widgets.Layout(width='250px'),
    style={'description_width': '90px'}
)

dropdown.observe(lambda change: show_summary(change['new']), names='value')
display(dropdown, out)
show_summary('All')
```
